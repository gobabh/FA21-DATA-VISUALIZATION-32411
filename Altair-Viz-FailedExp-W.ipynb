```python
import pandas as pd
import altair as alt
from altair.expr import datum, substring

# saving data into a file rather than embedding into the chart
alt.data_transformers.enable('json')

#alt.renderers.enable('notebook')
# alt.renderers.enable('jupyterlab')
alt.renderers.enable('default')

csv_path = 'eqData.csv'

try:
    eqDataDF = pd.read_csv(csv_path)
except:
    print("Dataset URL is not correct or not defined:")

eqDataDF = eqDataDF.dropna()
eqDataDF.head()
```

```python
alt.Chart(eqDataDF, background='lightgrey').transform_calculate(
    'Magnitude >=', substring(datum.Mweq, 0, 1)
).encode(
    longitude='Longitude:Q',
    latitude='Latitude:Q',
    color='Magnitude >=:N',
    tooltip=['Year:N','Latitude:N', 'Longitude:Q', 'Month:Q', 'Day:Q', 'Mweq:Q']
).mark_circle(size=15).project(
    type='mercator'
).properties(
    width=700,
    height=400,
)
```
